```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import cartopy.crs as ccrs
import seaborn as sns
```

```python
nprof = 30 #Specify a profile to plot
plt.plot(ds.PSAL[nprof], ds.PRES[nprof])

plt.xlabel('Temperature (C)')
plt.ylabel('Pressure (dbar)')
plt.title('Argo Profile from %s' % ds.PLATFORM_NUMBER[nprof].values)

plt.gca().invert_yaxis() #Flip the y-axis
```

```python
data=ds
```

```python
# Subplot example
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

nprof = 10 # Fist profile
ax1.plot(data.TEMP[nprof], data.PRES[nprof])
ax2.plot(data.PSAL[nprof], data.PRES[nprof])

nprof = 15 # Middle-ish profile
ax1.plot(data.TEMP[nprof], data.PRES[nprof])
ax2.plot(data.PSAL[nprof], data.PRES[nprof])

nprof = 25 # Last profile
ax1.plot(data.TEMP[nprof], data.PRES[nprof])
ax2.plot(data.PSAL[nprof], data.PRES[nprof])

ax1.set_ylabel('Pressure (dbar)')
ax1.set_xlabel('Temperature (C)')
ax2.set_xlabel('Salinity')
ax1.invert_yaxis()
ax1.legend()

# Add some gridlines
ax1.grid()
ax2.grid()

# Add a super title
fig.suptitle('Argo Float #%d' % data.PLATFORM_NUMBER[nprof].values, fontweight='bold', fontsize=16);
```

```python
nprof = 25 #Selected profile
plt.scatter(data.PSAL[nprof], data.TEMP[nprof])
plt.xlabel('Salinity')
plt.ylabel('Temperature (°C)')
plt.grid()

plt.title('Argo Float #%d' % data.PLATFORM_NUMBER[nprof].values, fontweight='bold');
```

```python
plt.figure(figsize=(8,6))

nprof = 25 #Selected profile
plt.scatter(data.PSAL[nprof], data.TEMP[nprof], c=data.PRES[nprof], cmap='viridis_r')
plt.xlabel('Salinity');
plt.ylabel('Temperature (°C)')

cbh = plt.colorbar();
cbh.set_label('Pressure (dbar)')
```
